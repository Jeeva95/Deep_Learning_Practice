```python
import keras
import matplotlib.pyplot as plt
import numpy as np
```

```python
def predict_img(img_path, model) :
  img = keras.utils.load_img(img_path, target_size = (256, 256))

  plt.imshow(img)
  plt.axis("off")
  plt.show()

  img_array = keras.utils.img_to_array(img)
  img_array = img_array / 255.0
  img_array = np.expand_dims(img_array, axis = 0)

  prediction = model.predict(img_array, verbose = 0)

  if prediction > 0.5 :
    print("Predicted : Dog", prediction[0][0] * 100)
  else :
    print("Prediction : Cat", (1 - prediction[0][0]) * 100)
```

```python
filename = list(uploaded.keys())[0]
print(filename)
```

```python
predict_img(f"/content/{filename}",model)
```
